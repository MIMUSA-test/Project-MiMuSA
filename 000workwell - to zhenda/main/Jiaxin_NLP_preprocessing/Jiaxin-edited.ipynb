```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('initial data as well as targeted output-RA-data-example -jiaxin-simple.csv')  #import reviews
df
```

```python
import re

def newtext(text):
    text = text.strip()
    text = text.replace('/\s\s+/g', ' ') # replace multiple spaces with a single space
    text = re.sub ('\s+', ' ', text)
    text = re.sub('@[^\s]+','',text)  # delete the username
    text = re.sub('&[^\s]+','',text)
    text = re.sub('#[^\s]+','',text)
    text = re.sub('".*?"', '', text)  # delete anything in quotation marks
    text = re.sub('http[s]?://\S+', '', text) # delete urls
    
    text = text.replace("extension service","extension-service") ## use replace words
    text = text.replace("extracurricular activity","extracurricular-activity")
    text = text.replace("alcohol addiction","alcohol-addiction")
    text = text.replace("bad condition","bad-condition")
    text = text.replace("health hazard","health-hazard")
    text = text.replace("old age","old-age")
    text = text.replace("persistent cruelty","persistent-cruelty")
    text = text.replace("poor shape","poor-shape")
    text = text.replace("sinking spell","sinking-spell")
    text = text.replace("Cut off","cut-off")
    text = text.replace("Cut away","cut-away")
    text = text.replace("Cut back","cut-back")
    text = text.replace("Fever has gone down","fever-gonedown")
    text = text.replace("take action","take-action")
    text = text.replace("advance directive","advance-directive")
    text = text.replace("all hail","all-hail")
    text = text.replace("care delivery","care-delivery")
    text = text.replace("electrolyte balance","electrolyte-balance")
    text = text.replace("fat farm","fat-farm")
    text = text.replace("feel like a million","feel-likeamillion")
    text = text.replace("feel like a million dollars","feel-likeamilliondollars")
    
    text = re.sub("\S*@\S*\s?",'',text)   # delete email address
    text = text.replace('\n', ' ').replace('\r', '')  # Clean up all "\n"

    text = re.sub(r'[\?\.\!]+(?=[\?\.\!])', '', text) ## replace mutiple ?, !, . with the last puntuation appeared 
    text = re.sub(r'[()@#&]', '', text)  #remove multiple non common puntuations
    
    numbers = '0123456789' ## add a space behind all puntuations except for the '.' decimal point 
    symbols = "?!,;:$" ## symbols to add in a black space behind 
    new_text = ''
    for index, char in enumerate(text):
      if char == ".": # check before and after
        try:
          if (not (text[index-1]) in numbers) or (not text[index+1] in numbers):
            new_text += ". "
          else:
            new_text += "."
        except IndexError:
          pass
      elif char in symbols:
        try:
          if text[index+1] != " ":
            if (text[index+1]) not in numbers:
              new_text += (char + " ")
            else:
              new_text += char
          else:
            new_text += char
        except IndexError:
          new_text += char
      else:
        new_text += char    
    text = new_text
    
    text= re.sub(' +', ' ', text)
    #text= re.sub(':', '', text)
    text= re.sub("[:']", '', text)
    #text = re.sub ('\s+', '', text)
    text = text.strip()
    return text.lower()
```

```python
df["Text"] = df['Text'].apply(lambda text: newtext(text))
```

```python
df.drop(df.columns[0], axis=1)
```
